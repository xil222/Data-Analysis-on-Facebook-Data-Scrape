```python
repos = []
repos.extend((statsdata1,statsdata2,statsdata3,statsdata4,statsdata5,statsdata6,statsdata7,statsdata8,statsdata9,statsdata10,statsdata11,statsdata12,statsdata13,statsdata14,statsdata15,statsdata16,statsdata17,statsdata18,statsdata19,statsdata20,statsdata21,statsdata22,statsdata23,statsdata24,statsdata25,statsdata26,statsdata27,statsdata28,statsdata29,statsdata30))
repos
```

```python
res = []
def ratio(repos):
    for x in repos:
        key = x.keys()
        most_value = x['local_info']    #shape (N, 7)  N is number of elements
        people = most_value[:,0]  #extract（students,faculty）（数字1student,2faculty）
        count1 = count2 = 0
        for x in people:
            if x == 1:
                count1+=1
            elif x==2:
                count2+=1
        res.append(count1/float(count2))
    
ratio(repos)
```
